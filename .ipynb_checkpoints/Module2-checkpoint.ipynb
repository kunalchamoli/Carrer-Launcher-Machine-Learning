```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('week2.csv', index_col= 'Trade_Date')
df1.head(10)
```

```python
df1.drop(['Date'], axis=1, inplace= True)
```

```python
df1.head(5)
```

```python
f, ax = plt.subplots(figsize = (25,15))
sns.pointplot(x = df1.index, y = df1['Close Price'] , color = 'blue')
plt.xticks(rotation = 45)
plt.grid()
ax.set(xlabel = 'dates' , ylabel = 'Close Price', title = 'Close Price VS Dates')
```

```python
plt.figure(figsize= (25, 15))
plt.stem(df1.index , df1['Day_Perc_Change'], '-')
plt.xticks(rotation = 45)
plt.grid()
plt.xlabel = 'Date'
plt.ylabel = 'Day Percentage Change'
```

```python
plt.figure(figsize= (25, 15))
plt.stem(df1.index , df1['Total Traded Quantity'], '-')
plt.xticks(rotation = 45)
plt.grid()
plt.xlabel = 'Date'
plt.ylabel = 'Day Percentage Change'
```

```python
df1['Date'] = df1.index
```

```python
df2 = df1.groupby('Trend').count()
```

```python
type(df2)
```

```python
df2
```

```python
df2.index
```

```python
plt.figure(figsize = (20, 10))
plt.pie(df2["Date"], labels = df2.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

```python
df3 = df1.groupby('Trend', as_index= False)['Total Traded Quantity'].median()
```

```python
df3.head()
```

```python
plt.figure(figsize= (20, 10))
plt.bar(df3['Trend'], df3['Total Traded Quantity'])
plt.xticks(rotation = 45, fontsize = 10)
plt.show()
```

```python
df4 = df1.groupby('Trend', as_index= False)['Total Traded Quantity'].mean()
```

```python
plt.figure(figsize= (20, 10))
plt.bar(df4['Trend'], df4['Total Traded Quantity'])
plt.xlabel("Trend")
plt.ylabel('Average Value')
plt.xticks(rotation = 45, fontsize = 10)
plt.show()
```

```python
plt.figure(figsize= (20, 10))
plt.hist(df1['% Dly Qt to Traded Qty'])
plt.title('Daily return (percentage) distribution')
plt.show()
```
