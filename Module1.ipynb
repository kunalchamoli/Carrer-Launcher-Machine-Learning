## Understanding Columns
**Prev_close** = Previous close refers to the prior day's value of a stock

**Open_price** = The price of the first trade for any listed stock is its daily opening price.Typically, a security's opening price is not identical to its prior day closing price. The difference is because after-hours trading has changed investor valuations or expectations for the security.

**Last_price** = The last price is the price on which most charts are based. The chart updates with each change of the last price, The last price is the most recent transaction, but it doesn't always accurately represent the price you would get if you were to buy or sell right now. The last price might have taken place at the bid or ask, or the bid or ask price might have changed as a result of or since the last price.

Share **Turnover** is a measure of stock liquidity calculated by dividing the total number of shares traded over a period by the average number of shares outstanding for the period. The higher the share turnover, the more liquid company shares are. total number of outstanding shares, which are shares that have been issued to investors and are available for purchase

**Total Traded Quantity** as a standalone metric only throws light on how many shares were traded in a day. To have a more informed view on how a stock has performed, it is critical to note the deliverable quantity for the stock.

**Deliverables** are the specific section of total volume traded shares which are actually transferred from one person's to another's demat account. In short, ownership is transferred. Volume traded = Intra day traded shares + Inter day traded share. Deliverables = Inter day traded share. Deliverables % = (Inter day traded shares/Total volume traded) Ã— 100 So what does it show? In other words, how it's helpful to investors? Increase in deliverable % shows a strong sign that long term investors are interested (entering) in this particular stock. (Because inter day trades are increasing and intra day is decreasing) Decrease in deliverable % shows a weak sign that long term investors are not interested (exiting) in this particular stock. (Because inter day trades are decreasing and intra day is increasing).

```python
def average(series, n):
    return series[-n:].mean()
```

```python
def Percent(series, n):
    return ((series[-1] - series[-n])/ series[-1])
```

```python
time = [7, 14, 30, 90,  180, 365]
for t in time:
    print("Average for {} days is {}".format(t,average(df['Average Price'],t)))
    print("Percent Profit/Loss for {} days is {}". format(t, Percent(df["Close Price"], t)))
```

```python
df["Day_Perc_Change"] = df["Close Price"].pct_change()*100
df
```

```python
df.fillna(0, inplace= True)
```

# 1.7

```python
condition = [
    (df['Day_Perc_Change'] > -0.5 ) & (df["Day_Perc_Change"] < 0.5),
    (df['Day_Perc_Change'] > 0.5 ) & (df["Day_Perc_Change"] < 1),
    (df['Day_Perc_Change'] > -1 ) & (df["Day_Perc_Change"] < -0.5),
    (df['Day_Perc_Change'] > 1 ) & (df["Day_Perc_Change"] < 3),
    (df['Day_Perc_Change'] > -3 ) & (df["Day_Perc_Change"] < -1),
    (df['Day_Perc_Change'] > 3) & (df["Day_Perc_Change"] < 7), 
    (df['Day_Perc_Change'] > -7 ) & (df["Day_Perc_Change"] < -3), 
    (df['Day_Perc_Change'] > 7 )
]
choices = ['No Change', 'Slight Positive' , 'Slight Negative', 'Positive', 'Negative',
          'Among Top Gainers', 'Among Top Losers', 'Bull run']
df['Trend'] = np.select(condition, choices, default= 'Bear Drop')
df
```
